```python
import bs4
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url='https://www.ebay.com/sch/i.html?_from=R40&_trksid=p2380057.m570.l1313&_nkw=drones&_sacat=0'
```

```python
page=urllib.request.urlopen(url)
```

```python
page_soup=BeautifulSoup(page,'html.parser')
print(page_soup)
```

```python
img_items=page_soup.find('div',{'class':'srp-river srp-layout-inner'})
```

```python
omg_div=img_items.find_all(class_='s-item__image-section')
```

```python
i=1
for img in omg_div:
    img_tag=img.find('img')
    img_src=img_tag.get('src')

    if img_src[:1]=='/':
        image='https:'+img_src
    else:
        image=img_src
    filename=str(i)
    i+=1
    img_file=open((filename+'.jpeg'),'wb')
    img_file.write(urllib.request.urlopen(image).read())  
    img_file.close()   
```
