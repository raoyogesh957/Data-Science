```python
import requests
from bs4 import BeautifulSoup
import urllib
import pandas as pd
#df=pd.DataFrame()

def get_soup(url):
    user='Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
    while True:
      link=requests.get(url,headers={'User-Agent':user})
      if link.status_code==200:
        soup=BeautifulSoup(link.content,'html.parser')
        break
    return soup

def get_links_page(soup):
    se=set()
    xyz=soup.find_all('div',{'class':'_2kHMtA'})
    for w in xyz:
        e=w.find('a',{'class':"_1fQZEK"}).get('href')
        se.add('https://flipkart.com'+e)
    se=list(se)
    print(len(se))
    return se

def get_data(url):
    while True:
      try:
        s=get_soup(url)
        a=[]
        b=[]
        a.append('Brand')
        b.append(s.title.text.strip())
        a.append('Price')
        b.append(s.find('div',{'class':'_30jeq3 _16Jk6d'}).text)
        for j in s.find_all('table',{'class':'_14cfVK'}):
            for z in j.find_all('tr',{'class':'_1s_Smc row'}):
                for x ,y in zip(z.find('td') , z.find('li')):
                    #print(x.text,":",y.text)
                    a.append(x.text)
                    b.append(y.text)
        df1=pd.DataFrame([b],columns=[a])
        break
      except:
          pass  
    return df1    
   

url='https://www.flipkart.com/search?q=laptop+under+30000+rs&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_13_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop+under+30000+rs%7CLaptops&requestId=8d25ec84-9036-4a72-8774-949e13731846&as-backfill=on'
c=0
links=set()
while True:
  try:
    soup=get_soup(url)
    clas = '_1LKTO3'
    abc=soup.find_all('a', {'class': clas})
    for i in abc:
      if i.text=='Next':
        next_link=i.get('href')
    links.add(url)
    print(url)
    if next_link is not None:
      url = 'https://flipkart.com'+next_link
      c+=1
      print('Link : ',c)
      se= get_links_page(soup)
      for z in se:
          df1=get_data(z)
          df=pd.concat([df,df1],axis=0,ignore_index=True)
  except:
    pass
  if c==9:
    break
```
