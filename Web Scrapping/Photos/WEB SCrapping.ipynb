```python
import bs4
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
```

## Images 

```python
url='https://www.amazon.in/deal/0031791c?showVariations=true&pf_rd_r=CMQQ6XBVHPNWN9KF9MYB&pf_rd_t=Events&pf_rd_i=deals&pf_rd_p=93f4fb92-379a-4f4e-b015-648c7e856116&pf_rd_s=slot-15&ref=dlx_deals_gd_dcl_img_0_0031791c_dt_sl15_16'
```

```python
page=requests.get(url,headers=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}))
```

```python
page_soup=BeautifulSoup(page.content,'html.parser')
print(page_soup)
```

```python
img_items=page_soup.find('div',{'class':'a-fixed-left-grid-col a-col-right'})
```

```python
img_items
```

```python
omg_div=img_items.find_all(class_='a-list-normal a-spacing-none no-margin-right overflow-hidden octopus-response-li-width')
```

```python
i=1
for img in omg_div:
    img_tag=img.find('img')
    img_src=img_tag.get('src')

    if img_src[:1]=='/':
        image='https:'+img_src
    else:
        image=img_src
    filename=str(i)
    i+=1
    img_file=open((filename+'.jpeg'),'wb')
    img_file.write(urllib.request.urlopen(image).read())  
    img_file.close()   
```
